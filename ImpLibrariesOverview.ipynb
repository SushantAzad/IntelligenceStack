```python
import json

"""Converting the json string to python object using loads"""
data = '{"name":"sushant","age":31}'
py_obj = json.loads(data)
print(py_obj)
print(py_obj['name'])
print(type(py_obj))

"""Converting the python object back to the json string"""
py_str = json.dumps(py_obj)
print(py_str)
print(type(py_str))
print(py_str[0:5])

"""We use the load, dump to do the same thing but with the file"""
```
