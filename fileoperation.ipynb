# File operations 

```python
"""This is for reading a file we get the error (FileNotFound) when the file is not present"""
with open("example.txt",'r') as file:
    # for line in file:
        # print(line.strip())
    content= file.read()
    print(type(content))
    print(content)

"""We use the file.read when we need to get the content of the file as whole"""
"""When we need to read the content line by line we use the for loop"""
```

```python
with open("example.txt" ,"w") as file:
    file.write("hello world \n");
    file.write("this is the new line"); 
"""when we use the w as the mode we overwrite the file. we use the append mode 'a' to add the content to the file without overwriting."""

```

```python
"""When we want to add multiple lines in the file"""
lines = ["\nfirst line \n", "second line\n", "third line \n"]
with open("example.txt", 'a') as file:
    file.writelines(lines)

```

## Working with binary files

```python
data = b"\x00\x01\x02\x03\x04"
with open("example.bin", 'wb') as file:
    file.write(data)
with open("example.bin", 'rb') as file:
    content = file.read()
    print(content)
```

- w+ mode is used to open the file to both reading and writing. If the file does not exists it will be created. the content will be overwritten
- We use file.seek() when we want to read a file after writing some content to it to move the cursor at the begining of the file
