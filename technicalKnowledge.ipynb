| Parameter               | Purpose                             |
| ----------------------- | ----------------------------------- |
| `model`                 | Model to use                        |
| `messages`              | Conversation history                |
| `temperature`           | Creativity/randomness (0 to 2)      |
| `top_p`                 | Alternative to temperature          |
| `max_completion_tokens` | Maximum output tokens               |
| `frequency_penalty`     | Reduces repeated phrases            |
| `presence_penalty`      | Encourages new topics               |
| `stop`                  | Stops generation on matching text   |
| `n`                     | Number of responses generated       |
| `stream`                | Stream tokens as they are generated |
| `user`                  | End user identifier                 |

## Most common parameters to use are 

| Parameter               | When to use                                |
| ----------------------- | ------------------------------------------ |
| `model`                 | Always                                     |
| `messages`              | Always                                     |
| `temperature`           | Control creativity                         |
| `max_completion_tokens` | Limit response length                      |
| `stream`                | For real time token streaming in chat apps |
| `response_format`       | When you need structured JSON output       |
| `tools`                 | When using function calling/tool calling   |
| `tool_choice`           | Force or control tool usage                |
