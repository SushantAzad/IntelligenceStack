### Charbot conversation with gradio UI


```python
import os
from dotenv import load_dotenv
from IPython.display import display, Markdown
from litellm import completion
from openai import OpenAI

load_dotenv(override=True)
```

```python
gemini_api_key = os.getenv("GOOGLE_API_KEY")
cerebras_api_key = os.getenv("CEREBRAS_API_KEY")
groq_api_key = os.getenv("GROQ_API_KEY")
```

```python
system_message = """
You're a helpful assistant
"""
```

```python
def chat(message, history):
    history = [{"role": h["role"], "content": h["content"]} for h in history]
    messages = (
        [{"role": "system", "content": system_message}]
        + history
        + [{"role": "user", "content": message}]
    )
    response = completion(model="gemini/gemini-2.5-flash", api_key=gemini_api_key,messages=messages, stream= True)
    stream =""
    for chunk in response:
        stream+=chunk.choices[0].delta.content or '' 
        yield stream
```

```python
import gradio as gr
view = gr.ChatInterface(fn=chat).launch(inbrowser=True,share=True)
```
