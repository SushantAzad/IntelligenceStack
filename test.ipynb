```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from scraper import fetch_website_links, fetch_website_contents
from IPython.display import Markdown, display
load_dotenv(override= True)
```

```python
links = fetch_website_links("https://www.geeksforgeeks.org/artificial-intelligence/aiml-introduction/")
links
```

```python
system_prompt = """You are a senior marketing strategist, content auditor, and information architect.

Your task is to analyze a list of website URLs and identify which links contain information that would be valuable for creating a professional marketing brochure.

For each URL, determine whether it is:

1. HIGHLY RELEVANT
   - Company overview
   - About us
   - Products
   - Services
   - Solutions
   - Features
   - Industries served
   - Case studies
   - Success stories
   - Testimonials
   - Benefits
   - Value proposition
   - Mission and vision
   - Leadership information
   - Certifications
   - Awards
   - Partnerships
   - Key differentiators

2. POSSIBLY RELEVANT
   - Blog articles
   - News updates
   - Resource pages
   - Whitepapers
   - Events
   - FAQs containing product or company information

3. NOT RELEVANT
   - Login pages
   - Account pages
   - Privacy policy
   - Terms and conditions
   - Cookie policy
   - Careers
   - Job listings
   - Support tickets
   - Documentation not useful for marketing
   - Search results
   - Category archives
   - Tag pages
   - Legal pages
   - Sitemap pages

Output the results in the following format:

## Highly Relevant Links
| URL | Reason |
|-----|--------|
| URL | Explanation |

## Possibly Relevant Links
| URL | Reason |
|-----|--------|
| URL | Explanation |

## Not Relevant Links
| URL | Reason |
|-----|--------|
| URL | Explanation |

After categorization, provide:

### Recommended Brochure Source Links
Rank the top links that should be reviewed first when creating the brochure.

### Information Coverage Analysis
Identify whether the available links likely contain:
- Company Overview
- Products/Services
- Benefits
- Customer Success Stories
- Testimonials
- Industry Expertise
- Contact Information

Highlight any missing brochure critical information."""
```

```python
user_prompt = f""" 
I want to create a professional marketing brochure.

Analyze the following URLs and categorize them according to their usefulness for brochure creation.

Links:
{links}


Requirements:
1. Categorize each link as Highly Relevant, Possibly Relevant, or Not Relevant.
2. Explain why each link belongs in that category.
3. Rank the top brochure worthy links from most valuable to least valuable.
4. Identify any content gaps that may require additional research.
5. Focus only on information useful for brochure creation.
"""
```

```python
gemini_base_url = "https://generativelanguage.googleapis.com/v1beta/openai/"
gemini_api_key = os.getenv("GOOGLE_API_KEY")
gemini = OpenAI(base_url=gemini_base_url, api_key= gemini_api_key)
responses = gemini.chat.completions.create(model = "gemini-2.5-flash", messages = [{"role":"user", "content":f"{user_prompt}"},{"role":"system", "content": f"{system_prompt}"}])
print(responses.choices[0].message.content)
links1 = responses.choices[0].message.content
```
