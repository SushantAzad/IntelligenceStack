## Making 2 chatbot argue


```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from litellm import Completions
load_dotenv(override=True)
gemini_api_key = os.getenv("GOOGLE_API_KEY")
```

```python
gemini_model = "gemini/gemini-2.5-flash"
ollama_model = "ollama/qwen3:8b"

gemini_system = "You are a chatbot who is very argumentative; \
you disagree with anything in the conversation and you challenge everything, in a snarky way."

ollama_system = "You are a very polite, courteous chatbot. You try to agree with \
everything the other person says, or find common ground. If the other person is argumentative, \
you try to calm them down and keep chatting."

gemini_msg = ["Hi there!"]
ollama_msg = ["Hi"]
```

```python
def gemini_call():
    messages = [{"role":"system",'content':gemini_system}]
    for gemini,ollama in zip(gemini_msg,ollama_msg):
        messages.append({"role":"assistant", "content":gemini})
        messages.append({"role":"user","content":ollama})
    response = completion(model=gemini_model, api_key=gemini_api_key ,messages=messages)

    display(Markdown(response.choices[0].message.content))
    return response.choices[0].message.content
```

```python
gemini_call()
```

```python
def ollama_call():
    messages = [{"role":"system","content":ollama_system }]
    for gemini,ollama in zip(gemini_msg,ollama_msg):
        messages.append({"role":"assistant","content":ollama})
        messages.append({"role":"user","content":gemini})
    messages.append({"role": "user", "content": gemini[-1]})
    response=completion(model=ollama_model,messages=messages)
    display(Markdown(response.choices[0].message.content))
    return response.choices[0].message.content
```

```python
ollama_call()
```

```python
gemini_msg = ["Hi there!"]
ollama_msg = ["Hi"]

display(Markdown(f"### Gemini:\n{gemini_msg[0]}\n"))
display(Markdown(f"### Ollama:\n{ollama_msg[0]}\n"))
for i in range(5):
    gemini_next = gemini_call()
    display(Markdown(f"### gemini:\n{gemini_next}\n"))
    gemini_msg.append(gemini_next)
    ollama_next = ollama_call()
    display(Markdown(f"### Ollama:\n{ollama_next}\n"))
    ollama_msg.append(ollama_next)
    
```
