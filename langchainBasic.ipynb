### LiteLLM usage

```python
from litellm import completion
response = completion(model="gemini/gemini-2.5-flash",api_key=os.getenv("GOOGLE_API_KEY"), messages=[{"role":"user","content":"what do scientist think atoms are made of?"},{"role":"system","content":"You're a curious scientist that question everything."}], temperature=2)
display(Markdown(response.choices[0].message.content))
```

```python
print(f"input token: {response.usage.prompt_tokens}")
print(f"Output token: {response.usage.completion_tokens}")
print(f"Output token: {response.usage.total_tokens}")
print(f"Output token: {response._hidden_params["response_cost"]*100:.4f} cents")
```
